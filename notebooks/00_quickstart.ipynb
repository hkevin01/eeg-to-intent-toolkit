# EEG-to-Intent Toolkit Quickstart

This notebook shows how to load a dummy dataset and run a tiny training loop with the CLI modules.

```python
# Imports
import numpy as np
import torch
from eegintent.data.loaders import load_dummy_dataset
from eegintent.training.trainer import make_trainer, make_dataloaders
```

```python
# Load dummy data
ds = load_dummy_dataset(n_samples=32, n_channels=8, n_times=128)
X = torch.tensor(ds.X)
y = torch.tensor(ds.y)
```

```python
# Make trainer and run one epoch
trainer, lit = make_trainer({"n_channels": 8, "n_classes": 2, "max_epochs": 1})
train_loader, val_loader = make_dataloaders(X, y, batch_size=8)
trainer.fit(lit, train_loader, val_loader)
```
