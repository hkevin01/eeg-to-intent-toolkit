# Personalization and Few-Shot Calibration
This tutorial shows how to use FiLM-based personalization and compute prototypical centroids.

```python
# Imports and seed
import torch
from eegintent.personalization.film import FiLM
from eegintent.personalization.proto import compute_prototypes, prototypical_loss
torch.manual_seed(0)
```

```python
# FiLM personalization over synthetic input
B, C, T, S = 4, 8, 128, 3
x = torch.randn(B, C, T)
sids = torch.randint(0, S, (B,))
film = FiLM(n_subjects=S, feat_dim=C)
y = film(x, sids)
y.shape
```

```python
# Prototypical networks loss demo
emb = torch.randn(12, 32)
labels = torch.randint(0, 3, (12,))
protos = compute_prototypes(emb, labels)
loss = prototypical_loss(emb, labels, protos)
float(loss.item())
```
